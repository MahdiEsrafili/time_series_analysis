```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.preprocessing import MinMaxScaler
from statsmodels.tsa.stattools import grangercausalitytests
%matplotlib inline
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)
```

```python
data = pd.read_csv('data/owid-covid-data.csv')
```

```python
data.columns
```

```python
data.drop(['new_tests', 'total_tests',
       'total_tests_per_thousand', 'new_tests_per_thousand',
       'new_tests_smoothed', 'new_tests_smoothed_per_thousand',
       'tests_per_case', 'positive_rate', 'tests_units', 'female_smokers',
       'male_smokers'], inplace=True, axis = 1)
```

```python
data.head()
```

```python
ir_data = data[data.location=='Iran']
ir_data.date = pd.to_datetime(ir_data.date) 
ir_data.set_index('date', inplace=True)
```

```python
ir_data.new_deaths.plot(figsize=(16,5))
```

```python
ir_data.loc[(ir_data.index > pd.to_datetime('2020-04-01')) & (ir_data.new_deaths==0), 'new_deaths'] = np.nan
ir_data.new_deaths.fillna(method='ffill', inplace=True)
```

```python
ir_data.new_deaths.plot(figsize=(16,5))
```

```python
ir_data = ir_data.asfreq('d')
iran_decompose = seasonal_decompose(ir_data.new_deaths)
```

```python
iran_decompose.plot()
plt.show()
```

```python
uk_data = data[data.location=='United Kingdom']
uk_data.date = pd.to_datetime(uk_data.date)
uk_data.set_index('date', inplace=True) 
uk_data = uk_data.asfreq('d')
uk_data.new_deaths.fillna(method= 'ffill', inplace =True)
uk_decomp = seasonal_decompose(uk_data.new_deaths)
```

```python
canada_data = data[data.location=='Canada']
canada_data.date = pd.to_datetime(canada_data.date)
canada_data.set_index('date', inplace=True) 
canada_data = canada_data.asfreq('d')
canada_data.new_deaths.fillna(method= 'ffill', inplace =True)
cnda_decomp = seasonal_decompose(canada_data.new_deaths)
```

```python
iran_nc_decomp.trend.plot(figsize=(16,5))
us_nc_decomp.trend.plot()
uk_nc_decomp.trend.plot()
cnda_nc_decomp.trend.plot()
plt.legend(['Iran', 'US','UK','Canada'])
for month in range(2,9):
    plt.axvline(pd.to_datetime('2020-0'+str(month)+'-01'), color='k', alpha=0.1, linestyle='--')
plt.show()
```

```python
min_max_scaler = MinMaxScaler()
```

```python
ir_death_trend = pd.DataFrame(iran_decompose.trend)
```

```python
ir_death_trend.head(2)
```

```python
ir_death_trend.new_deaths = min_max_scaler.fit_transform(iran_decompose.trend.values.reshape((-1,1)))[:,0]
```

```python
ir_death_trend.head()
```

```python
gsearch_iran['count'] = min_max_scaler.fit_transform((gsearch_iran.values.reshape((-1,1))))[:,0]
```

```python
plt.figure(figsize=(16,5))
plt.plot(ir_death_trend)
plt.plot(gsearch_iran)
plt.legend(['Iran Daily Deaths', 'Iran Googling Crona'])
plt.grid()
plt.show()
```

```python
print(gsearch_iran.values[:-2].shape)
print(gsearch_iran.isna().sum())
```

```python
print(ir_death_trend.values.shape)
print(ir_death_trend.dropna().shape)
```

```python
np.correlate(gsearch_iran.values[:-2].reshape((-1)), ir_death_trend.dropna().values.reshape((-1)))
```

```python

from dtw import *
alignment = dtw(gsearch_iran.values[:-2].reshape((-1)), ir_death_trend.dropna().values.reshape((-1)), keep_internals=True)

## Display the warping curve, i.e. the alignment curve
alignment.plot(type="threeway")
print(alignment.distance)

## Align and plot with the Rabiner-Juang type VI-c unsmoothed recursion
align2 = dtw(gsearch_iran.values[:-2].reshape((-1)), ir_death_trend.dropna().values.reshape((-1)), keep_internals=True, 
    step_pattern=rabinerJuangStepPattern(6, "c"))
align2.plot(type="twoway",offset=-2)
print(align2.distance)

```

```python
iran_smelling = pd.read_csv('data/iran-smelling.csv')
iran_smelling.Day = pd.to_datetime(iran_smelling.Day)
iran_smelling.set_index('Day', inplace = True)
iran_smelling.head()
```

```python
iran_smelling_trend.dropna(inplace=True)
```

```python
iran_smelling_trend['smelling'] = min_max_scaler.fit_transform((iran_smelling_trend.values.reshape((-1,1))))[:,0]
```

```python
iran_smelling_trend.shape
```

```python
plt.figure(figsize=(16,5))
plt.plot(ir_death_trend)
plt.plot(gsearch_iran)

plt.plot(iran_smelling_trend)
plt.legend(['Iran Daily Deaths', 'Iran Googling Crona', 'Iran Googling Smelling'])
plt.grid()
plt.show()
```

```python
test_df = iran_smelling_trend.join(ir_death_trend)
maxlag = 20
test = 'ssr_chi2test'
test_result = grangercausalitytests(test_df, maxlag=maxlag, verbose=False)
p_values = [round(test_result[i+1][0][test][1],4) for i in range(maxlag)]
min(p_values)
```
